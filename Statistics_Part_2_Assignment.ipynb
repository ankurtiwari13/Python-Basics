Ques -1. What is hypothesis testing in statistics?

-->  Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample of data to infer that a certain condition is true for the entire population. It's a way to make decisions about a population based on data from a sample. It involves setting up two competing hypotheses, the null hypothesis (H₀) and the alternative hypothesis (H₁), and then using statistical tests to determine which hypothesis is more likely to be true.

Ques -2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

-->  The null hypothesis (H₀) is a statement that there is no significant difference between two populations or variables. It is the hypothesis that the researcher is trying to disprove. The alternative hypothesis (H₁) is a statement that there is a significant difference between two populations or variables. It is the hypothesis that the researcher is trying to prove.

For example, if a researcher is testing the effectiveness of a new drug, the null hypothesis would be that the drug has no effect on the disease, while the alternative hypothesis would be that the drug does have an effect on the disease.

Ques -3. What is the significance level in hypothesis testing, and why is it important?

-->  The significance level (alpha or α) in hypothesis testing is the probability of rejecting the null hypothesis when it is actually true. It is the threshold at which we decide that the evidence against the null hypothesis is strong enough to reject it.

It is important because it helps us control the risk of making a Type I error, which is the error of rejecting a true null hypothesis. A lower significance level reduces the risk of a Type I error but increases the risk of a Type II error (failing to reject a false null hypothesis).
Commonly used significance levels are 0.05 (5%), 0.01 (1%), and 0.10 (10%). The choice of significance level depends on the specific research question and the consequences of making a Type I error.

Ques -4. What does a P-value represent in hypothesis testing?

-->  The P-value in hypothesis testing is the probability of obtaining test results at least as extreme as the observed results, assuming that the null hypothesis is true. A small P-value (typically less than the significance level) suggests that the observed data is unlikely to have occurred under the null hypothesis, leading to the rejection of the null hypothesis. A large P-value suggests that the observed data is consistent with the null hypothesis, and we fail to reject the null hypothesis.

Ques -5. How do you interpret the P-value in hypothesis testing?

--> The interpretation of the P-value is crucial in hypothesis testing. Here's how you interpret it:

*   **If the P-value is less than or equal to the significance level (α):** You reject the null hypothesis (H₀). This means that the observed data provides enough evidence to conclude that the alternative hypothesis (H₁) is likely true for the population. The results are considered statistically significant.
*   **If the P-value is greater than the significance level (α):** You fail to reject the null hypothesis (H₀). This means that the observed data does not provide enough evidence to conclude that the alternative hypothesis (H₁) is likely true for the population. The results are not considered statistically significant.



Ques -6. What are Type 1 and Type 2 errors in hypothesis testing?

--> In hypothesis testing, there are two main types of errors you can make:

*   **Type I Error (False Positive):** This occurs when you reject the null hypothesis (H₀) when it is actually true. In other words, you conclude there is a significant effect or difference when there isn't. The probability of making a Type I error is equal to the significance level (α).
*   **Type II Error (False Negative):** This occurs when you fail to reject the null hypothesis (H₀) when it is actually false. In other words, you conclude there is no significant effect or difference when there actually is. The probability of making a Type II error is denoted by β.

Here's a table to summarize:

| Decision             | H₀ is True        | H₀ is False       |
| :------------------- | :---------------- | :---------------- |
| **Reject H₀**        | Type I Error (α)  | Correct Decision  |
| **Fail to Reject H₀** | Correct Decision  | Type II Error (β) |



Ques -7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

--> The difference between a one-tailed and a two-tailed test in hypothesis testing lies in the direction of the alternative hypothesis (H₁).

*   **One-tailed test (or directional test):** This type of test is used when the alternative hypothesis specifies a direction of the effect or relationship. You are interested in whether the population parameter is either greater than or less than a certain value, but not both. The rejection region is entirely in one tail of the sampling distribution.
    *   Examples:
        *   H₀: The average test score is equal to 75.
        *   H₁: The average test score is *greater than* 75. (Right-tailed test)
        *   H₁: The average test score is *less than* 75. (Left-tailed test)

*   **Two-tailed test (or non-directional test):** This type of test is used when the alternative hypothesis does not specify a direction. You are interested in whether the population parameter is simply *different from* a certain value (either greater or less). The rejection region is split equally between both tails of the sampling distribution.
    *   Example:
        *   H₀: The average test score is equal to 75.
        *   H₁: The average test score is *not equal to* 75.



Ques -8. What is the Z-test, and when is it used in hypothesis testing?

--> The Z-test is a statistical hypothesis test that is used to determine whether there is a significant difference between the mean of a sample and the mean of a population, or between the means of two samples. It is used when the population standard deviation is known, or when the sample size is large (typically n > 30), allowing the sample standard deviation to be a good estimate of the population standard deviation.

The Z-test assumes that the data is normally distributed. The formula for the Z-test statistic is:

$$ Z = \frac{(\bar{x} - \mu)}{(\sigma / \sqrt{n})} $$

Where:
*   $\bar{x}$ is the sample mean
*   $\mu$ is the population mean
*   $\sigma$ is the population standard deviation
*   $n$ is the sample size

In a two-sample Z-test, the formula is slightly different and involves the means and standard deviations of both samples.

The Z-test is commonly used in various fields, including:

*   **Business:** To compare the average sales of two different marketing campaigns.
*   **Healthcare:** To compare the average effectiveness of two different treatments.
*   **Education:** To compare the average test scores of students from two different schools.

It's important to note that if the population standard deviation is unknown and the sample size is small, a t-test is generally more appropriate than a Z-test.

Ques -9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

--> The Z-score, also known as the standard score, is a measure of how many standard deviations a particular data point is away from the mean of a distribution. In hypothesis testing, the Z-score is calculated for a sample mean to determine how many standard errors the sample mean is away from the hypothesized population mean under the null hypothesis.

The formula for calculating the Z-score for a sample mean is:

$$ Z = \frac{(\bar{x} - \mu)}{(\sigma / \sqrt{n})} $$

Where:
*   $\bar{x}$ is the sample mean
*   $\mu$ is the hypothesized population mean (from the null hypothesis)
*   $\sigma$ is the population standard deviation
*   $n$ is the sample size

What the Z-score represents in hypothesis testing:

*   **Standardized Measure:** The Z-score standardizes the difference between the sample mean and the hypothesized population mean, making it comparable across different datasets with different scales.
*   **Distance from the Mean:** A higher absolute Z-score indicates that the sample mean is further away from the hypothesized population mean.
*   **Probability:** The Z-score is used in conjunction with the standard normal distribution (Z-distribution) to find the P-value. The P-value is the probability of observing a sample mean as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true. A large absolute Z-score corresponds to a small P-value, which provides evidence against the null hypothesis.
*   **Decision Making:** By comparing the calculated Z-score to critical Z-values (determined by the significance level), you can decide whether to reject or fail to reject the null hypothesis. If the calculated Z-score falls in the rejection region (beyond the critical Z-values), you reject the null hypothesis.



Ques -10. What is the T-distribution, and when should it be used instead of the normal distribution?

--> The T-distribution, also known as the Student's t-distribution, is a probability distribution that is used in hypothesis testing when the sample size is small and the population standard deviation is unknown. It is similar in shape to the normal distribution but has heavier tails, meaning it has more probability in the tails and less in the center. This accounts for the increased uncertainty when working with smaller sample sizes and an unknown population standard deviation.

**When to use the T-distribution instead of the normal distribution (Z-distribution):**

You should use the T-distribution when both of the following conditions are met:

1.  **The sample size (n) is small:** Generally, this means n < 30. However, this is a guideline, and the t-distribution is technically more accurate whenever the population standard deviation is unknown, regardless of sample size. The difference between the t-distribution and the normal distribution becomes negligible as the sample size increases.
2.  **The population standard deviation (σ) is unknown:** If you know the population standard deviation, you should use the Z-test and the normal distribution, regardless of the sample size.

**Key characteristics of the T-distribution:**

*   It is bell-shaped and symmetric, like the normal distribution.
*   It has a mean of zero.
*   It is defined by its degrees of freedom (df), which is typically calculated as n - 1 (where n is the sample size). The degrees of freedom determine the shape of the distribution; as the degrees of freedom increase, the t-distribution approaches the normal distribution.
*   It has heavier tails than the normal distribution, which means that extreme values are more likely to occur in a t-distribution.

Ques -12. What is the T-test, and how is it used in hypothesis testing?

--> The T-test is a statistical hypothesis test used to determine if there is a significant difference between the means of two groups, or if the mean of a single group is significantly different from a known or hypothesized population mean. It is particularly useful when the sample size is small and the population standard deviation is unknown.

**How the T-test is used in hypothesis testing:**

1.  **Formulate Hypotheses:**
    *   **Null Hypothesis (H₀):** States that there is no significant difference between the means being compared. For example, H₀: μ₁ = μ₂ (the means of two groups are equal) or H₀: μ = μ₀ (the sample mean is equal to a hypothesized population mean).
    *   **Alternative Hypothesis (H₁):** States that there is a significant difference between the means. This can be one-tailed (e.g., H₁: μ₁ > μ₂) or two-tailed (e.g., H₁: μ₁ ≠ μ₂).

2.  **Choose the Significance Level (α):** This is the probability of rejecting the null hypothesis when it is true (Type I error). Common values are 0.05 or 0.01.

3.  **Calculate the T-statistic:** The formula for the t-statistic varies depending on the type of t-test being conducted (one-sample, independent two-sample, paired two-sample). The general idea is to measure the difference between the observed sample mean(s) and the hypothesized population mean(s) relative to the variability within the sample(s).

    *   **One-Sample T-test:** Compares the mean of a single sample to a known or hypothesized population mean.
        $$ t = \frac{(\bar{x} - \mu_0)}{(s / \sqrt{n})} $$
        Where: $\bar{x}$ is the sample mean, $\mu_0$ is the hypothesized population mean, $s$ is the sample standard deviation, and $n$ is the sample size.

    *   **Independent Two-Sample T-test:** Compares the means of two independent groups. The formula depends on whether equal variances are assumed (pooled t-test) or not (Welch's t-test).

    *   **Paired Two-Sample T-test:** Compares the means of two related groups (e.g., measurements before and after an intervention on the same subjects). It works with the differences between pairs of observations.

4.  **Determine the Degrees of Freedom (df):** The degrees of freedom depend on the type of t-test and the sample size(s). For a one-sample t-test, df = n - 1.

5.  **Find the P-value:** Using the calculated t-statistic and the degrees of freedom, you find the probability of obtaining a t-statistic as extreme as, or more extreme than, the observed value under the null hypothesis. This is done using the T-distribution table or statistical software.

6.  **Make a Decision:**
    *   If the P-value is less than or equal to the significance level (α), you **reject the null hypothesis**. This suggests there is sufficient evidence to conclude a significant difference exists.
    *   If the P-value is greater than the significance level (α), you **fail to reject the null hypothesis**. This suggests there is not enough evidence to conclude a significant difference exists based on the sample data.



Ques -13. What is the relationship between Z-test and T-test in hypothesis testing?

--> The Z-test and T-test are both statistical tests used to compare means and make inferences about population parameters based on sample data. However, they are applicable under different conditions and have a specific relationship:

**The T-test can be seen as a generalization of the Z-test.**

Here's how they are related:

1.  **Under Ideal Conditions (Known Population Standard Deviation):** If you know the true population standard deviation ($\sigma$), you should use the Z-test. In this scenario, the sample mean's distribution is normal, and the Z-test is the appropriate test statistic.

2.  **When Population Standard Deviation is Unknown:** In most real-world scenarios, the population standard deviation is unknown. When this is the case, we have to estimate it using the sample standard deviation ($s$). Using the sample standard deviation introduces more variability and uncertainty, especially with smaller sample sizes.

3.  **The Role of the T-distribution:** Because we are using an *estimated* standard deviation (the sample standard deviation, $s$) instead of the true population standard deviation ($\sigma$), the distribution of the test statistic is not the standard normal distribution (Z-distribution) anymore, but rather the **T-distribution**. The T-distribution has heavier tails than the normal distribution to account for this added uncertainty from estimating $\sigma$.

4.  **Convergence with Increasing Sample Size:** As the sample size ($n$) increases, the sample standard deviation ($s$) becomes a more reliable estimate of the population standard deviation ($\sigma$). Consequently, the T-distribution with increasing degrees of freedom (n-1) approaches the standard normal distribution.

**In essence:**

*   The **Z-test** is used when you have the most information (known population standard deviation) or a very large sample size.
*   The **T-test** is used when you have less information (unknown population standard deviation) and particularly when your sample size is small.
*   The **T-test becomes more similar to the Z-test as the sample size grows**, because the T-distribution converges to the normal distribution.


Ques -14. What is a confidence interval, and how is it used to interpret statistical results?

-->  A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (like a mean or proportion) with a certain level of confidence. It’s a fundamental concept in inferential statistics used to express the uncertainty or precision of an estimate.

#How to Interpret It
A 95% confidence interval means that if you were to take 100 different samples and compute a CI for each, about 95 of them would contain the true population parameter.

It does not mean there's a 95% chance the true value is in your specific interval—once the interval is calculated, the true value either is or isn't in it.

Ques -15. What is the margin of error, and how does it affect the confidence interval?

--> The **margin of error** is a critical component of a confidence interval. It quantifies the amount of random sampling error in the results of a survey or experiment. In simpler terms, it's the "plus or minus" figure often reported with survey results, indicating the range within which the true population value is likely to fall.


**How the Margin of Error Affects the Confidence Interval:**

The margin of error directly determines the width of the confidence interval. The confidence interval is calculated as:

$$ \text{Confidence Interval} = \text{Sample Statistic} \pm \text{Margin of Error} $$

*   **Larger Margin of Error:** Leads to a **wider** confidence interval. A wider interval indicates greater uncertainty about the true population parameter. This can happen due to a higher confidence level (you need a wider range to be more confident), a smaller sample size (more variability), or greater variability in the population.
*   **Smaller Margin of Error:** Leads to a **narrower** confidence interval. A narrower interval indicates less uncertainty and a more precise estimate of the true population parameter. This can be achieved with a lower confidence level, a larger sample size, or less variability in the population.

Ques -16. How is Bayes' Theorem used in statistics, and what is its significance?

-->
**How Bayes' Theorem is Used in Statistics:**

Bayes' Theorem is the cornerstone of Bayesian statistics and is used in various ways:

1.  **Updating Beliefs:** The core idea is to start with a prior belief about the probability of a hypothesis ($P(H)$) and then update that belief to a posterior probability ($P(H|E)$) after observing new data (evidence E). This iterative process allows for the incorporation of new information as it becomes available.

2.  **Parameter Estimation:** In Bayesian inference, instead of estimating a single point value for a parameter (like in frequentist methods), we estimate a probability distribution for the parameter. Bayes' Theorem is used to update the prior distribution of a parameter based on the likelihood of the observed data, resulting in a posterior distribution. This posterior distribution represents our updated belief about the plausible values of the parameter.

3.  **Model Comparison:** Bayes' Theorem can be used to compare different statistical models by calculating the probability of each model given the observed data. This is often done using Bayes Factors, which are ratios of marginal likelihoods.

4.  **Making Predictions:** Bayesian methods use the posterior distribution of parameters to make predictions about future observations. These predictions are often expressed as predictive distributions, which account for the uncertainty in the parameter estimates.

5.  **Spam Filtering:** A classic example is using Bayes' Theorem to classify emails as spam or not spam based on the presence of certain words. The prior probability is the overall probability of an email being spam, the likelihood is the probability of seeing certain words in spam/non-spam emails, and the posterior probability is the probability of an email being spam given the words it contains.

**Significance of Bayes' Theorem:**

The significance of Bayes' Theorem and Bayesian statistics lies in several aspects:

1.  **Incorporating Prior Knowledge:** It provides a formal way to incorporate prior knowledge or beliefs into the statistical analysis, which can be particularly useful when data is scarce or expensive to obtain.
2.  **Intuitive Interpretation:** The posterior probability $P(H|E)$ has a more intuitive interpretation than the P-value in frequentist statistics. It directly tells us the probability of the hypothesis being true given the observed data.
3.  **Quantifying Uncertainty:** Bayesian methods naturally provide measures of uncertainty in the form of probability distributions for parameters and predictions.
4.  **Flexibility:** Bayesian methods can be very flexible and applied to a wide range of complex statistical models.

Ques -18. What is the Chi-square goodness of fit test, and how is it applied?

--> The **Chi-square goodness-of-fit test** is a statistical hypothesis test used to determine if an observed frequency distribution of a single categorical variable differs significantly from an expected frequency distribution. In simpler terms, it helps you assess whether your sample data "fits" a hypothesized distribution for the population.

**How is it applied (Steps):**

1.  **Formulate Hypotheses:**
    *   **Null Hypothesis (H₀):** The observed frequency distribution is the same as the expected frequency distribution (i.e., there is no significant difference between the observed and expected distributions).
    *   **Alternative Hypothesis (H₁):** The observed frequency distribution is different from the expected frequency distribution (i.e., there is a significant difference).

2.  **Choose a Significance Level (α):** This is the probability of rejecting the null hypothesis when it is true (Type I error). Common values are 0.05 or 0.01.

3.  **Calculate the Expected Frequencies:** For each category, calculate the expected frequency under the assumption that the null hypothesis is true. This is typically done by multiplying the total sample size by the expected proportion for that category.

    $$ \text{Expected Frequency}_i = \text{Total Sample Size} \times \text{Expected Proportion}_i $$

4.  **Calculate the Chi-square Test Statistic ($\chi^2$):** This statistic measures the discrepancy between the observed frequencies and the expected frequencies. The formula is:

    $$ \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} $$

    Where:
    *   $O_i$ is the observed frequency for category $i$.
    *   $E_i$ is the expected frequency for category $i$.
    *   The summation is across all categories.

5.  **Determine the Degrees of Freedom (df):** The degrees of freedom for the Chi-square goodness-of-fit test are calculated as:

    $$ df = k - 1 $$

    Where $k$ is the number of categories in the variable.

6.  **Find the P-value:** Using the calculated $\chi^2$ test statistic and the degrees of freedom, you find the probability of obtaining a Chi-square value as extreme as, or more extreme than, the observed value under the null hypothesis. This is done using the Chi-square distribution table or statistical software.

7.  **Make a Decision:**
    *   If the P-value is less than or equal to the significance level (α), you **reject the null hypothesis**. This suggests there is sufficient evidence to conclude that the observed distribution is significantly different from the expected distribution.
    *   If the P-value is greater than the significance level (α), you **fail to reject the null hypothesis**. This suggests there is not enough evidence to conclude that the observed distribution is significantly different from the expected distribution, meaning the observed data "fits" the hypothesized distribution.

**Assumptions of the Chi-square Goodness-of-Fit Test:**

*   **Categorical Data:** The data must be from a single categorical variable.
*   **Independent Observations:** The observations must be independent of each other.
*   **Expected Frequencies:** The expected frequency for each category should ideally be at least 5. If some expected frequencies are too small, the test results may not be reliable, and you might need to combine categories.



Ques -19. What is the F-distribution, and when is it used in hypothesis testing?

--> The **F-distribution**, also known as the Snedecor's F-distribution or the Fisher–Snedecor distribution, is a continuous probability distribution that arises in the testing of whether two independent samples have the same variance or in comparing means of three or more groups. It is primarily used in **Analysis of Variance (ANOVA)** and in testing the equality of variances.

**When is the F-distribution Used in Hypothesis Testing?**

The F-distribution is most commonly used in the following scenarios:

1.  **Analysis of Variance (ANOVA):** This is the most common application of the F-distribution. ANOVA is used to test if there are significant differences between the means of three or more independent groups. The F-statistic in ANOVA is the ratio of the variance between the groups to the variance within the groups. If the F-statistic is large, it suggests that the differences between group means are larger than what would be expected by random chance, leading to the rejection of the null hypothesis (that all group means are equal).

    *   **One-Way ANOVA:** Used to compare the means of three or more independent groups based on one categorical independent variable.
    *   **Two-Way ANOVA:** Used to compare the means of groups classified by two independent categorical variables, and to assess the interaction effect between the two variables.

2.  **Testing the Equality of Variances:** The F-test can be used to test whether the variances of two populations are equal. This is often done as a preliminary step before conducting a t-test for independent samples, as the choice of t-test formula (pooled or Welch's) depends on whether the variances are equal.

3.  **Regression Analysis:** In regression analysis, the F-test is used to assess the overall significance of the regression model. It tests whether at least one of the predictor variables in the model has a statistically significant linear relationship with the response variable.



Ques -20. What is an ANOVA test, and what are its assumptions?

--> **ANOVA (Analysis of Variance)** is a statistical test used to determine if there are any statistically significant differences between the means of three or more independent (unrelated) groups. It is an extension of the t-test, which is used to compare the means of only two groups.

Instead of performing multiple t-tests (which increases the risk of a Type I error), ANOVA uses the F-distribution to compare the variability *between* the group means to the variability *within* the groups.

**The core idea behind ANOVA:**

If the variability between the group means is significantly larger than the variability within the groups, then it is likely that at least one of the group means is different from the others.

**Assumptions of ANOVA:**

ANOVA is a parametric test, meaning it makes certain assumptions about the data. These assumptions should ideally be met for the results of the ANOVA test to be valid and reliable:

1.  **Independence of Observations:** The observations within each group and between the groups must be independent. This means that the value of one observation does not influence the value of another observation. This is often ensured through proper study design and random sampling.

2.  **Normality:** The data within each group should be approximately normally distributed. ANOVA is relatively robust to violations of this assumption, especially with larger sample sizes, due to the Central Limit Theorem. However, severe departures from normality can affect the results. You can check this assumption using visual methods (histograms, Q-Q plots) or statistical tests (e.g., Shapiro-Wilk test).

3.  **Homogeneity of Variances (Homoscedasticity):** The variances of the dependent variable should be approximately equal across all groups. This means that the spread of the data within each group should be similar. Violations of this assumption (heteroscedasticity) can lead to inflated Type I error rates. You can check this assumption using tests like Levene's test or Bartlett's test. If this assumption is violated, you might need to use a non-parametric alternative (like the Kruskal-Wallis test) or a modified ANOVA test (like Welch's ANOVA).

4.  **Continuous Dependent Variable:** The dependent variable (the variable you are measuring) must be continuous (interval or ratio scale).

Ques -22. What is the F-test, and how does it relate to hypothesis testing?

--> The **F-test** is a statistical test that uses the F-distribution to determine if the variances of two populations are equal or if there are significant differences between the means of three or more groups (as in ANOVA). It's a crucial part of hypothesis testing when comparing variances or analyzing multiple group means.

**How the F-test Relates to Hypothesis Testing:**

In hypothesis testing, the F-test is used to evaluate hypotheses that involve comparing variances or means across multiple groups. Here's the general process:

1.  **Formulate Hypotheses:**
    *   **For comparing two variances:**
        *   H₀: The variances of the two populations are equal ($\sigma_1^2 = \sigma_2^2$).
        *   H₁: The variances of the two populations are not equal ($\sigma_1^2 \neq \sigma_2^2$) (two-tailed) or one variance is greater than the other (one-tailed).
    *   **For ANOVA (comparing means of 3+ groups):**
        *   H₀: All group means are equal ($\mu_1 = \mu_2 = \dots = \mu_k$).
        *   H₁: At least one group mean is different from the others.

2.  **Choose a Significance Level (α):** This is the probability of rejecting the null hypothesis when it is true.

3.  **Calculate the F-statistic:** The F-statistic is calculated as a ratio of variances.
    *   **For comparing two variances:** The F-statistic is the ratio of the larger sample variance to the smaller sample variance ($F = s_1^2 / s_2^2$).
    *   **For ANOVA:** The F-statistic is the ratio of the variance between groups to the variance within groups ($F = \text{Between-Group Variance} / \text{Within-Group Variance}$).

4.  **Determine the Degrees of Freedom:** The F-test has two sets of degrees of freedom: numerator degrees of freedom and denominator degrees of freedom, which depend on the specific test being conducted (e.g., sample sizes for variance comparison, number of groups and total sample size for ANOVA).

5.  **Find the P-value:** Using the calculated F-statistic and the degrees of freedom, you determine the probability of observing an F-statistic as extreme as, or more extreme than, the calculated value under the null hypothesis. This is done using the F-distribution table or statistical software.

6.  **Make a Decision:**
    *   If the P-value is less than or equal to the significance level (α), you **reject the null hypothesis**. This suggests there is sufficient evidence to support the alternative hypothesis (i.e., variances are unequal or at least one group mean is different).
    *   If the P-value is greater than the significance level (α), you **fail to reject the null hypothesis**. This suggests there is not enough evidence to conclude that the variances are unequal or that the group means are different.

```python
#Ques -3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(k, p):
  """
  Calculates the probability distribution function (PDF) of a Bernoulli distribution.

  Args:
    k: The outcome (0 for failure, 1 for success).
    p: The probability of success (between 0 and 1).

  Returns:
    The probability of the given outcome k.
  """
  if k == 1:
    return p
  elif k == 0:
    return 1 - p
  else:
    return 0 # For invalid inputs of k

# Example usage:
p_success = 0.7
outcome1 = 1
outcome2 = 0
outcome3 = 2

print(f"PDF for k={outcome1} with p={p_success}: {bernoulli_pdf(outcome1, p_success)}")
print(f"PDF for k={outcome2} with p={p_success}: {bernoulli_pdf(outcome2, p_success)}")
print(f"PDF for k={outcome3} with p={p_success}: {bernoulli_pdf(outcome3, p_success)}")
```

```python
# Ques -4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Define parameters for the binomial distribution
n = 10  # number of trials
p = 0.5 # probability of success in each trial
size = 1000 # number of times to repeat the experiment (simulate)

# Simulate the binomial distribution
# This generates 'size' number of outcomes, where each outcome is the number of successes in 'n' trials
binomial_sample = binom.rvs(n, p, size=size)

# Plot the histogram of the simulated data
plt.figure(figsize=(10, 6))
plt.hist(binomial_sample, bins=np.arange(-0.5, n + 1.5, 1), density=True, alpha=0.7, color='skyblue', edgecolor='black')

# Plot the theoretical PMF for comparison
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)
plt.stem(x, pmf, linefmt='red', markerfmt='ro', basefmt=' ')
plt.vlines(x, 0, pmf, color='red', linestyles='-', lw=1)


plt.title(f"Simulated Binomial Distribution (n={n}, p={p}) Histogram with Theoretical PMF")
plt.xlabel("Number of Successes")
plt.ylabel("Probability / Frequency")
plt.xticks(np.arange(0, n + 1))
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Ques -9 (Practical). Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def calculate_and_plot_zscores(data):
  """
  Calculates the Z-scores for each data point in a dataset and plots them.

  Args:
    data: A list or NumPy array of numerical data.
  """
  # Calculate Z-scores
  z_scores = zscore(data)

  # Print the calculated Z-scores
  print("Original Data:", data)
  print("Calculated Z-scores:", z_scores)

  # Plot the Z-scores
  plt.figure(figsize=(10, 6))
  plt.stem(range(len(data)), z_scores, linefmt='b-', markerfmt='bo', basefmt=' ')
  plt.axhline(0, color='gray', linestyle='--', linewidth=0.8) # Add a horizontal line at Z=0
  plt.title("Z-scores of the Dataset")
  plt.xlabel("Data Point Index")
  plt.ylabel("Z-score")
  plt.grid(axis='y', linestyle='--', alpha=0.6)
  plt.show()

# Example Usage:
# Create a sample dataset
sample_data = np.array([10, 12, 12, 13, 12, 11, 14, 13, 15, 10, 10, 12, 14, 13, 11])

# Calculate and plot Z-scores
calculate_and_plot_zscores(sample_data)

# Another example with different data
sample_data_2 = np.array([5, 8, 9, 12, 6, 7, 10, 11, 15, 4])
calculate_and_plot_zscores(sample_data_2)
```

```python
# Ques -10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon # Using an Exponential distribution as a non-normal example

# Define parameters for the original non-normal distribution (Exponential)
lambda_param = 0.5 # Rate parameter for Exponential distribution
# The mean and standard deviation of an Exponential distribution are 1/lambda

# Parameters for the simulation
sample_sizes = [1, 5, 30, 100] # Different sample sizes to demonstrate CLT
num_samples = 1000 # Number of samples to draw for each sample size

# Create a figure to hold the plots
plt.figure(figsize=(15, 10))

# Plot the original distribution for reference
original_data = expon.rvs(scale=1/lambda_param, size=10000) # Generate large sample for original distribution
plt.subplot(2, 2, 1)
plt.hist(original_data, bins=50, density=True, alpha=0.7, color='lightgreen', edgecolor='black')
x = np.linspace(original_data.min(), original_data.max(), 100)
plt.plot(x, expon.pdf(x, scale=1/lambda_param), 'r-', lw=2, label='Exponential PDF')
plt.title(f"Original Distribution (Exponential, λ={lambda_param})")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)


# Simulate and plot the distribution of sample means for different sample sizes
for i, n in enumerate(sample_sizes):
    sample_means = []
    for _ in range(num_samples):
        # Draw a random sample of size 'n' from the exponential distribution
        sample = expon.rvs(scale=1/lambda_param, size=n)
        # Calculate the mean of the sample
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    # Plot the histogram of sample means
    plt.subplot(2, 2, i + 1) # Plotting starts from subplot 1
    if i == 0: # Already plotted original distribution in subplot 1
        continue
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')

    # Overlay a normal distribution for comparison (based on CLT)
    # Mean of sample means approaches population mean (1/lambda)
    # Standard deviation of sample means (standard error) approaches population std dev / sqrt(n)
    # For Exponential: population std dev = 1/lambda
    clt_mean = 1 / lambda_param
    clt_std_err = (1 / lambda_param) / np.sqrt(n)
    xmin, xmax = plt.xlim()
    x_norm = np.linspace(xmin, xmax, 100)
    p_norm = norm.pdf(x_norm, clt_mean, clt_std_err)
    plt.plot(x_norm, p_norm, 'k-', lw=2, label='Normal PDF (CLT)')


    plt.title(f"Distribution of Sample Means (n={n})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
#Ques -13. Generate random variables and calculate their corresponding probabilities using the binomial distribution.
import numpy as np
from scipy.stats import binom

# Define parameters for the binomial distribution
n_trials = 20 # number of trials
p_success = 0.6 # probability of success in each trial

# Generate a few random variables from the binomial distribution
num_random_variables = 10
random_variables = binom.rvs(n=n_trials, p=p_success, size=num_random_variables)

print(f"Generated random variables from Binomial(n={n_trials}, p={p_success}):")
print(random_variables)

print("\nCalculating probabilities (PMF) for these outcomes:")
# Calculate the probability (PMF) for each generated random variable
for rv in random_variables:
  probability = binom.pmf(k=rv, n=n_trials, p=p_success)
  print(f"P(X = {rv}) = {probability:.6f}")

# You can also calculate the probability for a specific outcome you are interested in
specific_outcome = 12
probability_specific = binom.pmf(k=specific_outcome, n=n_trials, p=p_success)
print(f"\nProbability of exactly {specific_outcome} successes: P(X = {specific_outcome}) = {probability_specific:.6f}")

# Calculate the cumulative probability (CDF) for a specific value
cdf_value = 10
cumulative_probability = binom.cdf(k=cdf_value, n=n_trials, p=p_success)
print(f"Probability of {cdf_value} or fewer successes: P(X <= {cdf_value}) = {cumulative_probability:.6f}")

# Calculate the probability of a range of values (e.g., between 8 and 14 successes inclusive)
lower_bound = 8
upper_bound = 14
# P(8 <= X <= 14) = P(X <= 14) - P(X <= 7)
probability_range = binom.cdf(k=upper_bound, n=n_trials, p=p_success) - binom.cdf(k=lower_bound - 1, n=n_trials, p=p_success)
print(f"Probability of between {lower_bound} and {upper_bound} successes: P({lower_bound} <= X <= {upper_bound}) = {probability_range:.6f}")
```

```python
#Ques -14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_and_plot_single_zscore(data, data_point):
    """
    Calculates the Z-score for a specific data point within a dataset
    and plots its position on a standard normal distribution.

    Args:
      data: A list or NumPy array of numerical data (the sample).
      data_point: The specific data point for which to calculate the Z-score.
    """
    # Calculate the mean and standard deviation of the dataset
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate the Z-score for the given data point
    z_score = (data_point - mean) / std_dev

    print(f"Dataset Mean: {mean:.4f}")
    print(f"Dataset Standard Deviation: {std_dev:.4f}")
    print(f"Data Point: {data_point}")
    print(f"Calculated Z-score for {data_point}: {z_score:.4f}")

    # Plot the standard normal distribution
    plt.figure(figsize=(10, 6))
    x = np.linspace(-4, 4, 100) # Range for standard normal distribution
    plt.plot(x, norm.pdf(x, 0, 1), label='Standard Normal Distribution (Z-distribution)')

    # Plot the position of the calculated Z-score
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
    plt.plot(z_score, norm.pdf(z_score, 0, 1), 'ro') # Mark the point on the curve

    plt.title(f"Z-score of {data_point} in relation to the Standard Normal Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

# Example Usage:
# Create a sample dataset
sample_data = np.array([10, 12, 12, 13, 12, 11, 14, 13, 15, 10, 10, 12, 14, 13, 11])

# Choose a data point from the dataset
specific_point = 15

# Calculate and plot the Z-score for the specific data point
calculate_and_plot_single_zscore(sample_data, specific_point)

# Another example with a different data point
specific_point_2 = 10
calculate_and_plot_single_zscore(sample_data, specific_point_2)

# Example with a data point not in the original sample but within the distribution's range
specific_point_3 = 13.5
calculate_and_plot_single_zscore(sample_data, specific_point_3)
```

```python
# Ques -15. Implement hypothesis testing using Z-statistics for a sample dataset (One-Sample Z-test).
import numpy as np
from statsmodels.stats.weightstats import ztest

# --- Scenario ---
# A manufacturing company claims that the average weight of their product is 100 grams.
# A quality control manager takes a random sample of 30 products and finds the sample mean weight.
# The population standard deviation is known to be 5 grams.
# We want to test if the average weight is significantly different from 100 grams at a 0.05 significance level.

# --- Given Data ---
sample_data = np.array([102, 98, 101, 105, 99, 100, 103, 97, 100, 101,
                        104, 99, 102, 100, 103, 98, 101, 102, 100, 104,
                        99, 101, 102, 100, 103, 98, 101, 102, 100, 104]) # Sample data
population_mean_h0 = 100  # Hypothesized population mean under the null hypothesis (μ₀)
population_std_dev = 5    # Known population standard deviation (σ)
alpha = 0.05              # Significance level (α)
# Sample size (n) can be calculated from the sample data
sample_size = len(sample_data)
sample_mean = np.mean(sample_data)

print(f"Sample Size: {sample_size}")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Population Standard Deviation (known): {population_std_dev}")
print(f"Hypothesized Population Mean (H₀): {population_mean_h0}")
print(f"Significance Level (α): {alpha}")
print("-" * 30)

# --- Hypothesis Formulation ---
# Null Hypothesis (H₀): The average weight is equal to 100 grams (μ = 100).
# Alternative Hypothesis (H₁): The average weight is not equal to 100 grams (μ ≠ 100). (Two-tailed test)

# --- Perform One-Sample Z-test ---
# ztest function from statsmodels performs the z-test
# value: The hypothesized population mean (μ₀)
# ddof=0: Degrees of freedom adjustment for standard deviation calculation (use 0 for population std dev)
# alternative: 'two-sided', 'larger', 'smaller' based on H₁
z_statistic, p_value = ztest(x1=sample_data, value=population_mean_h0, ddof=0, alternative='two-sided')

print(f"Calculated Z-statistic: {z_statistic:.4f}")
print(f"Calculated P-value: {p_value:.4f}")
print("-" * 30)

# --- Decision Rule ---
# If P-value <= alpha, reject the null hypothesis.
# If P-value > alpha, fail to reject the null hypothesis.

# --- Make a Decision ---
print("Decision:")
if p_value <= alpha:
    print(f"Since the P-value ({p_value:.4f}) is less than or equal to the significance level ({alpha}), we reject the null hypothesis (H₀).")
    print("Conclusion: There is sufficient evidence to suggest that the average weight of the product is significantly different from 100 grams.")
else:
    print(f"Since the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis (H₀).")
    print("Conclusion: There is not enough evidence to suggest that the average weight of the product is significantly different from 100 grams.")
```

```python
# Ques -16. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy.stats import t

# --- Scenario ---
# A researcher wants to estimate the average height of students in a school.
# They take a random sample of 25 students and measure their heights.
# We want to calculate a 95% confidence interval for the true average height of students in the school.
# We will use a t-distribution since the population standard deviation is unknown and the sample size is relatively small.

# --- Given Data ---
sample_heights = np.array([165, 170, 168, 172, 175, 167, 171, 169, 173, 170,
                           166, 174, 170, 169, 172, 171, 168, 170, 175, 173,
                           169, 172, 170, 167, 171]) # Sample data (heights in cm)
confidence_level = 0.95 # 95% confidence level

# Calculate sample statistics
sample_size = len(sample_heights)
sample_mean = np.mean(sample_heights)
sample_std_dev = np.std(sample_heights, ddof=1) # Use ddof=1 for sample standard deviation

print(f"Sample Size: {sample_size}")
print(f"Sample Mean: {sample_mean:.4f} cm")
print(f"Sample Standard Deviation: {sample_std_dev:.4f} cm")
print(f"Confidence Level: {confidence_level}")
print("-" * 30)

# --- Calculate the Confidence Interval ---

# 1. Determine the degrees of freedom
degrees_of_freedom = sample_size - 1

# 2. Find the critical t-value
# For a two-tailed confidence interval, we need alpha/2 in each tail.
alpha = 1 - confidence_level
t_critical = t.ppf(1 - alpha/2, degrees_of_freedom) # ppf is the inverse of the CDF

print(f"Degrees of Freedom: {degrees_of_freedom}")
print(f"Critical t-value: {t_critical:.4f}")

# 3. Calculate the margin of error
standard_error = sample_std_dev / np.sqrt(sample_size)
margin_of_error = t_critical * standard_error

print(f"Standard Error: {standard_error:.4f} cm")
print(f"Margin of Error: {margin_of_error:.4f} cm")

# 4. Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("-" * 30)
print(f"{int(confidence_level*100)}% Confidence Interval for the Mean Height: ({lower_bound:.4f} cm, {upper_bound:.4f} cm)")

# --- Interpretation ---
print("\n--- Interpretation ---")
print(f"We are {int(confidence_level*100)}% confident that the true average height of students in the school")
print(f"lies within the interval from {lower_bound:.4f} cm to {upper_bound:.4f} cm.")
print("\nThis means that if we were to take many random samples of the same size from this school")
print("and calculate a 95% confidence interval for each sample, approximately 95% of those intervals")
print("would contain the true population mean height.")
```

```python
# Ques -19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
import numpy as np
from scipy.stats import poisson

# Define the parameter for the Poisson distribution (lambda)
lambda_param = 3 # Average number of events

print(f"Poisson Distribution with lambda (λ) = {lambda_param}")
print("-" * 40)

# Calculate the CDF for specific values
# CDF(k) is the probability of observing k or fewer events: P(X <= k)

k1 = 2
cdf_k1 = poisson.cdf(k1, mu=lambda_param)
print(f"CDF for k = {k1}: P(X <= {k1}) = {cdf_k1:.6f}")
print(f"Interpretation: The probability of observing {k1} or fewer events is {cdf_k1:.6f}.")

k2 = 5
cdf_k2 = poisson.cdf(k2, mu=lambda_param)
print(f"\nCDF for k = {k2}: P(X <= {k2}) = {cdf_k2:.6f}")
print(f"Interpretation: The probability of observing {k2} or fewer events is {cdf_k2:.6f}.")

# Calculate the probability of observing exactly a certain number of events using CDF
# P(X = k) = P(X <= k) - P(X <= k-1)
k_exact = 4
cdf_k_exact = poisson.cdf(k_exact, mu=lambda_param)
cdf_k_minus_1 = poisson.cdf(k_exact - 1, mu=lambda_param)
pmf_k_exact = cdf_k_exact - cdf_k_minus_1
print(f"\nProbability of exactly k = {k_exact}: P(X = {k_exact})")
print(f"Calculated using CDF: P(X <= {k_exact}) - P(X <= {k_exact - 1}) = {cdf_k_exact:.6f} - {cdf_k_minus_1:.6f} = {pmf_k_exact:.6f}")
# Verify with PMF
pmf_check = poisson.pmf(k_exact, mu=lambda_param)
print(f"Calculated using PMF: P(X = {k_exact}) = {pmf_check:.6f}")


# Calculate the probability of observing more than a certain number of events
# P(X > k) = 1 - P(X <= k)
k_more_than = 3
prob_more_than_k = 1 - poisson.cdf(k_more_than, mu=lambda_param)
print(f"\nProbability of more than k = {k_more_than}: P(X > {k_more_than}) = 1 - P(X <= {k_more_than}) = {prob_more_than_k:.6f}")
print(f"Interpretation: The probability of observing more than {k_more_than} events is {prob_more_than_k:.6f}.")

# Calculate the probability of observing a value within a range
# P(a <= X <= b) = P(X <= b) - P(X <= a-1)
lower_bound = 2
upper_bound = 4
prob_range = poisson.cdf(upper_bound, mu=lambda_param) - poisson.cdf(lower_bound - 1, mu=lambda_param)
print(f"\nProbability of between {lower_bound} and {upper_bound} events (inclusive): P({lower_bound} <= X <= {upper_bound}) = {prob_range:.6f}")
print(f"Interpretation: The probability of observing between {lower_bound} and {upper_bound} events (inclusive) is {prob_range:.6f}.")
```

```python
# Ques -24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
  """
  Calculates the skewness and kurtosis of a dataset.

  Args:
    data: A list or NumPy array of numerical data.

  Returns:
    A tuple containing the skewness and kurtosis of the dataset.
  """
  # Calculate skewness
  # Skewness measures the asymmetry of the distribution.
  # For a perfectly normal distribution, skewness is 0.
  data_skewness = skew(data)

  # Calculate kurtosis
  # Kurtosis measures the "tailedness" of the distribution.
  # For a standard normal distribution, kurtosis (excess kurtosis) is 0.
  # Positive kurtosis indicates heavier tails (leptokurtic), negative indicates lighter tails (platykurtic).
  data_kurtosis = kurtosis(data) # By default, scipy calculates excess kurtosis

  return data_skewness, data_kurtosis

# Example Usage:
# Define parameters for a normal distribution
mean = 0
std_dev = 1
size = 500 # Number of data points

# Generate a dataset from a normal distribution
np.random.seed(42) # for reproducibility
normal_data = np.random.normal(loc=mean, scale=std_dev, size=size)

print(f"Generated data from Normal(μ={mean}, σ={std_dev}) with size = {size}")
print("-" * 30)

# Calculate skewness and kurtosis
calculated_skewness, calculated_kurtosis = calculate_skewness_kurtosis(normal_data)

print(f"Calculated Skewness: {calculated_skewness:.4f}")
print(f"Calculated Kurtosis (Excess): {calculated_kurtosis:.4f}")

print("\n--- Interpretation for a Normal Distribution ---")
print("For a perfectly normal distribution:")
print("- Skewness is 0.")
print("- Kurtosis (Excess) is 0.")
print("\nOur calculated values are close to 0, which is expected for a dataset generated from a normal distribution,")
print("though there will be some variation due to random sampling.")

# Example with a skewed distribution (e.g., Exponential)
from scipy.stats import expon
lambda_param = 0.5
exponential_data = expon.rvs(scale=1/lambda_param, size=size)
exp_skewness, exp_kurtosis = calculate_skewness_kurtosis(exponential_data)

print("\n--- Interpretation for an Exponential Distribution ---")
print(f"Generated data from Exponential(λ={lambda_param}) with size = {size}")
print(f"Calculated Skewness: {exp_skewness:.4f} (Expected > 0 for right-skewed)")
print(f"Calculated Kurtosis (Excess): {exp_kurtosis:.4f} (Expected > 0 for leptokurtic)")
print("\nThese values are significantly different from 0, as expected for a non-normal distribution like the Exponential distribution.")
```

```python
#Ques -1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
#interpret the results.
import numpy as np
from scipy.stats import norm

# Sample data
sample = [52, 55, 48, 50, 53, 49, 51, 54, 50, 52]
sample_mean = np.mean(sample)
sample_size = len(sample)

# Known population parameters
population_mean = 50
population_std = 3  # Known population standard deviation

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.3f}")
print(f"P-Value: {p_value:.4f}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference between sample and population mean.")
```

```python
#Ques-3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
import numpy as np
from scipy.stats import norm

# Sample data (e.g., test scores)
sample_data = [72, 75, 78, 74, 77, 73, 76, 79, 74, 75]
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Known population parameters
population_mean = 70
population_std = 5  # Known population standard deviation

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.3f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference between sample and population mean.")
```

```python
#Ques-4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data
sample_data = [102, 98, 100, 101, 99, 97, 103, 100, 98, 99]
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Known population parameters
population_mean = 100
population_std = 2  # Known population standard deviation

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.3f}")
print(f"P-Value: {p_value:.4f}")

# Decision threshold
alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)

# Interpretation
if abs(z_score) > z_critical:
    print("Result: Reject the null hypothesis.")
else:
    print("Result: Fail to reject the null hypothesis.")

# -------------------------------
# 📊 Visualization of Decision Region
# -------------------------------
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.5, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.5, label='Rejection Region (Right)')

# Plot the Z-score
plt.axvline(z_score, color='green', linestyle='--', linewidth=2, label=f'Z-score = {z_score:.2f}')

# Labels and legend
plt.title("Two-Tailed Z-Test: Decision Regions")
plt.xlabel("Z-score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#Ques-5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical Z-values for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)

    # Critical values in terms of the sample mean
    lower_crit = mu0 - z_critical * se
    upper_crit = mu0 + z_critical * se

    # X range for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions under H0 and H1
    y0 = norm.pdf(x, mu0, se)  # Null hypothesis
    y1 = norm.pdf(x, mu1, se)  # Alternative hypothesis

    # Plot the distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='H₀: μ = {}'.format(mu0), color='blue')
    plt.plot(x, y1, label='H₁: μ = {}'.format(mu1), color='green')

    # Shade Type I Error (α): Reject H₀ when it's true
    plt.fill_between(x, y0, where=(x < lower_crit) | (x > upper_crit), color='red', alpha=0.4, label='Type I Error (α)')
    plt.fill_between(x, y1, where=(x >= lower_crit) & (x <= upper_crit), color='orange', alpha=0.4, label='Type II Error (β)')

    # Plot critical values
    plt.axvline(lower_crit, color='black', linestyle='--', label='Critical Values')
    plt.axvline(upper_crit, color='black', linestyle='--')

    # Labels and legend
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example usage
visualize_type1_type2_errors()
```

```python
#Ques -6. Write a Python program to perform an independent T-test and interpret the results.
import numpy as np
from scipy.stats import ttest_ind

# Simulate two independent samples (e.g. test scores from two groups)
group1 = np.random.normal(loc=70, scale=10, size=50)  # Mean = 70
group2 = np.random.normal(loc=75, scale=10, size=50)  # Mean = 75
#Perfom independent two samples test
t_stat, p_value = ttest_ind(group1, group2)
# Print results
print(f"Group 1 Mean: {np.mean(group1):.2f}")
print(f"Group 2 Mean: {np.mean(group2):.2f}")
print(f"T-Statistic: {t_stat:.3f}")
print(f"P-Value: {p_value:.4f}")
# Interpret the result
if p_value < 0.05:
    print("Result: Reject the null hypothesis. There is a significant difference between the two groups.")
    print("Interpretation : There is a significant between the two group means")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference between the two groups.")
    print("Interpretation : There is no significant between the two group means")
```

```python
#Ques -7. Perform a paired sample T-test using Python and visualize the comparison results.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# Simulate paired data (e.g., before and after treatment scores)
np.random.seed(42)
before = np.random.normal(loc=70, scale=8, size=30)
after = before + np.random.normal(loc=2, scale=5, size=30)  # Slight improvement

# Perform paired t-test
t_stat, p_value = ttest_rel(before, after)

# Print results
print(f"Mean Before: {np.mean(before):.2f}")
print(f"Mean After: {np.mean(after):.2f}")
print(f"T-Statistic: {t_stat:.3f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Interpretation: There is a significant difference between the paired samples.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Interpretation: No significant difference between the paired samples.")

# -------------------------------
# 📊 Visualization
# -------------------------------
plt.figure(figsize=(10, 5))

# Plot paired values with lines
for i in range(len(before)):
    plt.plot([1, 2], [before[i], after[i]], marker='o', color='gray', alpha=0.6)

# Plot mean values
plt.plot(1, np.mean(before), 'ro', label='Mean Before')
plt.plot(2, np.mean(after), 'bo', label='Mean After')

# Customize plot
plt.xticks([1, 2], ['Before', 'After'])
plt.title("Paired Sample Comparison")
plt.ylabel("Score")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Ques -9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default is 0.95 for 95%)

    Returns:
        (lower_bound, upper_bound): Tuple with confidence interval bounds
    """
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_critical * std_err
    return mean - margin_of_error, mean + margin_of_error

# Example usage
sample_data = [52, 55, 48, 50, 53, 49, 51, 54, 50, 52]
ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Ques -10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default is 0.95 for 95%)

    Returns:
        float: Margin of error
    """
    n = len(data)
    sample_std = np.std(data, ddof=1)
    std_err = sample_std / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)
    moe = t_critical * std_err
    return moe

# Example usage
sample_data = [52, 55, 48, 50, 53, 49, 51, 54, 50, 52]
confidence_level = 0.95

moe = margin_of_error(sample_data, confidence=confidence_level)
print(f"Margin of Error at {int(confidence_level*100)}% confidence: ±{moe:.2f}")
```

```python
#Ques -11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior, likelihood, false_positive, prevalence_complement):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
        prior (float): P(Disease)
        likelihood (float): P(Positive | Disease)
        false_positive (float): P(Positive | No Disease)
        prevalence_complement (float): P(No Disease)

    Returns:
        float: P(Disease | Positive)
    """
    numerator = likelihood * prior
    denominator = (likelihood * prior) + (false_positive * prevalence_complement)
    posterior = numerator / denominator
    return posterior

# Given values
P_Disease = 0.01
P_Positive_given_Disease = 0.99
P_Positive_given_NoDisease = 0.05
P_NoDisease = 1 - P_Disease

# Calculate posterior
posterior = bayes_theorem(P_Disease, P_Positive_given_Disease, P_Positive_given_NoDisease, P_NoDisease)
print(f"Probability of having the disease given a positive test: {posterior:.4f}")
```

```python
#Ques -13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
#data.
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
# Define observed data (Contigency table)
data = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Smoker': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No']
})

# Create a contingency table
contingency_table = pd.crosstab(data['Gender'], data['Smoker'])
# Perform chi - square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
# Display Expected frequencies
print("Expected Frequencies:")
print(pd.DataFrame(expected, columns=contingency_table.columns, index=contingency_table.index).round(2))


```

```python
#Ques -14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

# Step 1: Define observed data
# Example: Observed frequencies of dice rolls (not perfectly fair)
observed = np.array([18, 22, 20, 16, 24, 20])  # Frequencies for faces 1–6

# Step 2: Define expected frequencies
# For a fair die, each face should appear equally often
expected = np.array([20, 20, 20, 20, 20, 20])  # Uniform distribution

# Step 3: Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Display results
print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected)
print(f"\nChi-Square Statistic: {chi2_stat:.3f}")
print(f"P-Value: {p_value:.4f}")

# Step 5: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Interpretation: The observed distribution significantly differs from the expected distribution.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Interpretation: The observed distribution does not significantly differ from the expected distribution.")
```
