Ques -21. What are the assumptions for applying the Central Limit Theorem?

-->
#Assumptions for Applying the Central Limit Theorem
#1. Independence of Observations
Each observation in the sample must be independent of the others.

- This means the outcome of one observation does not influence another.

- In practice, this is often satisfied by random sampling or random assignment.

> If sampling without replacement, the sample size should be less than 10% of the population to maintain independence.

#2. Identically Distributed Variables
The sampled variables should come from the same distribution (i.e., they are identically distributed).

- All observations should have the same mean and variance.

- This is typically true when sampling from a single population.

#3. Finite Mean and Variance
The population from which you're sampling must have a finite mean ğœ‡ and finite variance ğœ2.

- If the variance is infinite (as in some heavy-tailed distributions), the CLT may not apply.

#4. Sufficiently Large Sample Size
The sample size ğ‘› must be large enough for the sampling distribution of the mean to approximate normality.

- Rule of thumb: ğ‘› â‰¥ 30

- If the population is highly skewed or non-normal, a larger ğ‘› may be needed.
